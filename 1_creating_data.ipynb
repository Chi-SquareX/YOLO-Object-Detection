```
import json
import pandas as pd
import numpy as np
import re
```

```
f = open('/content/drive/MyDrive/Object_Detection/HQ_new_annotations.json') 
data = json.load(f)
[print(i) for i in data['img_1001_2_0.jpg46076']]
```

#func


```
df=pd.DataFrame(data=None)
for k, v in data.items():
  if isinstance(v, dict):
    for a,b in v.items():
        if a=='regions':
          new=[]
          for i in range(len(b)-1,-1,-1):
            new.append(b.pop().copy())
    for a,b in v.items():
        if a=='size':
          e=b
    for a,b in v.items():
        if a=='filename':
          f=b
  df_new=pd.DataFrame(new)
  df_new['filename']=f
  df_new['size']=e

  for i,j in df_new['region_attributes'].items():
    if isinstance(j, dict):
      for a,b in j.items():
        df_new.loc[i,'region']=b
  for i,j in df_new['shape_attributes'].items():
    if isinstance(j, dict):
      k=0
      for a,b in j.items():
        if k==0:
          df_new.loc[i,'name']=b
        if k==1:
          df_new.loc[i,'x_min']=min(b)
          df_new.loc[i,'x_max']=max(b)
        if k==2:
          df_new.loc[i,'y_min']=min(b)
          df_new.loc[i,'y_max']=max(b)
        k=k+1
  df1=df_new.drop(columns=['shape_attributes','region_attributes','name']).copy()
  df=pd.concat([df,df1])
  print(df1.head())
  print()
  #break
df.head()
```
